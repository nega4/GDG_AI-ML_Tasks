```python
import pandas as pd
df = pd.read_csv("Housing.csv")
df.head()
```

###**pd.read_csv()**
-opens the file and turns it into a DataFramesf(spreadsheet inside python).
###**df.head()**-
- is method that belongs to a DataFrame objects  that shows the first 5 rows of the dataset.


```python
df.columns
print("columns: ", df.columns.tolist())
```

###**df.columns.tolist()**
-helps us know what features we have and what target we want to know.

```python
y=df['price']
x=df.drop('price', axis=1)
```

Since we want to predict the price given all the other features of the house we store all the prices to **y** and we store everything else to **x**.

```python
x = pd.get_dummies(x, drop_first=True)
# x.head() #view what changed later
```

Since we have a categorical data(yes or no and furnished, unfernished or unfernished) with in the cells of the some columns we use the get_dummies() to change a totally numerical values.

```python
from sklearn.model_selection import train_test_split

# 80% training, 20% testing
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# Check shapes
print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

**LinearRegression()**
* this creates a new model object.*

**model.fit(x_train, y_train)**
* teaches the model.

```python

y_pred = model.predict(x_test)

print(y_pred[:10])
```

```python
# calculating errors
from sklearn.metrics import mean_squared_error, mean_absolute_error


mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
```

```python
# New house as a dictionary
new_house = {
    'area': 2000,
    'bedrooms': 4,
    'bathrooms': 3,
    'stories': 2,
    'mainroad': 'yes',
    'guestroom': 'no',
    'basement': 'yes',
    'hotwaterheating': 'no',
    'airconditioning': 'yes',
    'parking': 2,
    'prefarea': 'yes',
    'furnishingstatus': 'furnished'
}

# Convert to DataFrame
import pandas as pd
new_df = pd.DataFrame([new_house])

# Encode categorical columns like before
new_df = pd.get_dummies(new_df, drop_first=True)

# Align new_df with x_train columns (in case some columns are missing)
new_df = new_df.reindex(columns=x_train.columns, fill_value=0)

# Predict price
predicted_price = model.predict(new_df)
print("Predicted Price:", predicted_price[0])
```

In this case we use the model to predict the price of a house after we are given the values of the prices.
