# **TASK_2_(non-technical)**





**Step 1- Problem:**

 Predicting whether a person will show up for their doctor's appointment based on past behavior and demographic data.


# **Step 2 ML Type:**

Supervised Learning

This is a supervised learning problem because we will have historical data where we know whether people showed up or not.  
The model can learn patterns from these labeled examples and predict future attendance for new patients.


# **Step 3 ML Work Flow**
1. Collect data on past appointments (date, time, patient info, reminders sent, etc...)
2. Preprocess data (handle missing values, encode categorical features)
3. Split data into training and test sets
4. Train a supervised ML model (e.g., logistic regression, decision tree)
5. Evaluate the model (accuracy, precision, recall)
6. Deploy the model to predict future appointment attendance


# **Step 4 Possible Failure Point:**
Bad / insufficient data-because many patients donot go sick and as reslut there will be no enough data to predict the patterns of such individuals.

If the dataset does not cover enough examples of no-shows, the model may not learn the patterns correctly and give inaccurate predictions.  
**Fix:** Collect more historical data and ensure a balanced representation of show/no-show cases.
