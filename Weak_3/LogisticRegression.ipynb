## STEP 1:
* Import the required libraries





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## **STEP 2:**
*  Load the Breast Cancer dataset into a DataFrame


```python
from google.colab import files

uploaded = files.upload()
```

And oading  the CSV file into pandas DataFrame

```python

df = pd.read_csv("breast_cancer_bd.csv")
df.head() # Displaying the first 5 rows to check correct loading of the given dataset.
```


* Inspect the data

```python
print(df.shape)      # rows x columns
print(df.info())     # column types and missing data
print(df.describe()) # basic stats of numeric columns
```

## **Step 3:**


*   Handle Missing Values

```python
print(df.isnull().sum()) #checking which value has missing value
```

```python
# Fill missing values → df.fillna(value)
df['Bare Nuclei'] = df['Bare Nuclei'].replace('?', np.nan) # Replace '?' with NaN
df['Bare Nuclei'] = df['Bare Nuclei'].astype(float)   # Convert to float
median_value = df['Bare Nuclei'].median()   # Fill missing values safely
df['Bare Nuclei'] = df['Bare Nuclei'].fillna(median_value)
```

```python
print(df.isnull().sum())
```

## **Step 4:**
* Drop Non-Informative Columns

In the given dataset *Sample code number* is just an ID and has no predictive power so we have to drop it.


```python
df = df.drop(columns=['Sample code number'], errors='ignore') # Drop the 'Sample code number' column

# df.head(25) # Check the first 25 rows to confirm dropping of the column
```

##**Step 5:**
* Separating Features(x) and Targets(y)

From the given Dataset:
    
 * Features (X) -
Clump Thickness, Uniformity of Cell Size, Uniformity of Cell Shape, Marginal Adhesion, Single Epithelial Cell Size, Bare Nuclei, Bland Chromatin, Normal Nucleoli, Mitoses
  * Target (y) - Class (benign = 2, malignant = 4)

```python
X = df.drop(columns=['Class']) # Features (all columns except 'Class')

y = df['Class'] # Target
```

#**Step 6:**
* Spliting the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42) # Split: 70% training, 30% testing


#check shapes
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

note from the code:
* test_size=0.3 → 30% of data is for testing, 70% for training
* random_state=42 → ensures reproducible split
* X_train / y_train → what the model will learn from
* X_test / y_test → what the model will be evaluated on

#**Step 7:**
* Train the Logistic Regression Model

* ### Feature Scaling
  * for better performance using *StandardScaler* to make sure that numeric features are on a similar scale.


```python
scaler = StandardScaler() # Initialize scaler
X_train_scaled = scaler.fit_transform(X_train) # Fit scaler on training data and transform

```

```python
X_test_scaled = scaler.transform(X_test) # Transform test data (don’t fit again!)
```

* ### Train Logistic Regression

```python
model = LogisticRegression() # Initialize  and tran model

model.fit(X_train_scaled, y_train)
```

* ###  Make Predictions

```python
# Predict on test data and Check first 10 predictions
y_pred = model.predict(X_test_scaled)

print(y_pred[:10])
```

##**Step 8:**
* Evaluate Model Performance

* ## Accuracy Score
  * Measures the proportion of correct predictions

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

* ## Confusion Matrix

```python

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

* ## Classification Report

```python

report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

## Analysing the results:
  * ## Accuracy-
  Accuracy=(TP+TN)/(+FP+FN+TP+TN​ ).

 Since the code in the above cell, i.e. *print("Accuracy:", accuracy)*  print a value of  **Accuracy: 0.9666666666666667**, we can deduce that the model correctly predicted 96.67% of tumors.
 * ## Confusion Matrix-
 As we can see from the above result of the code *print("Confusion Matrix:\n", cm)* we can deduce:

 * Given :
    * TN = 141 → 141 benign tumors correctly identified as benign

    * FP = 2 → 2 benign tumors wrongly predicted as malignant  (false alarms)

    * FN = 5 → 5 malignant tumors wrongly predicted as benign (critical mistake)

    * TP = 62 → 62 malignant tumors correctly detected

The Logistic Regression model successfully classified tumors as benign or malignant with high accuracy, showing that simple supervised learning models can be effective for medical diagnosis when trained on clean and well-prepared data.
