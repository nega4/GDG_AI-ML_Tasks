**Objective:**  
- Create a list of 5 numbers  
- Convert it to a NumPy array  
- Calculate mean, maximum, and sum  
- Print the results clearly


```python
# Step 1: import NUmPy as np
import numpy as np

```

```python
# Step 2: create a list of 5 numbers
numbers=[1,2,3,4,5]
```

```python
# Step 3: convert list to NumPy array
arr= np.array(numbers)
```

```python
# Step 4: Calculate mean, max and sum
mean_val = np.mean(arr)
max_val = np.max(arr)
sum_val = np.sum(arr)
```

```python
# Step 5: print results clearly
print("Numbers:", arr)
print("Mean:", mean_val)
print("Maximum value:", max_val)
print("Sum:", sum_val)
```
